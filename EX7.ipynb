```python
# Install NLTK (if needed)
!pip install nltk

import nltk
import random
from nltk.corpus import brown
from nltk.tag import hmm

# Download datasets
nltk.download('brown')
nltk.download('universal_tagset')

# Convert corpus to LIST (IMPORTANT FIX)
tagged_sentences = list(brown.tagged_sents(tagset='universal'))

# Shuffle
random.shuffle(tagged_sentences)

# Split data
split = int(len(tagged_sentences) * 0.8)
train_data = tagged_sentences[:split]
test_data = tagged_sentences[split:]

# Train HMM model
trainer = hmm.HiddenMarkovModelTrainer()
model = trainer.train_supervised(train_data)

# Evaluate
accuracy = model.evaluate(test_data)
print("Model Accuracy:", round(accuracy * 100, 2), "%")

# Test sentence
sentence = "The quick brown fox jumps over the lazy dog".split()
print("\nTagged Output:")
print(model.tag(sentence))
```
