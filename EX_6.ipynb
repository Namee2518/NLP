```python
from collections import defaultdict
text = "I love NLP I love coding"

words = text.lower().split()

bigram_counts = defaultdict(int)
unigram_counts = defaultdict(int)
for i in range(len(words) - 1):
    unigram_counts[words[i]] += 1
    bigram = (words[i], words[i + 1])
    bigram_counts[bigram] += 1
unigram_counts[words[-1]] += 1
vocab_size = len(set(words))

def bigram_probability(w1, w2):
    bigram_count = bigram_counts[(w1, w2)]
    unigram_count = unigram_counts[w1]

    probability = (bigram_count + 1) / (unigram_count + vocab_size)
    return probability
prob = bigram_probability("love", "coding")
print("Probability of ('love','coding'):", prob)
```
